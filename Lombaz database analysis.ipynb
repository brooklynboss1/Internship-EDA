# Load Dataset and import Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
db = pd.read_csv("lomaz_db.csv")
db
```

# Businesses getting delivery requests from customers

```python
customers = db[db["delivery_requests"] == "yes"]
customers
```

# Not yet getting delivery requests 

```python
no_customers = db[db["delivery_requests"] == "no"]
no_customers
```

# Visualize delivery request count

```python
yes = db[db['delivery_requests'] == 'yes'].count()
number_yes = yes['delivery_requests']
number_yes
```

```python
no = db[db['delivery_requests'] == 'no'].count()
number_no = no['delivery_requests']
number_no
```

```python
sizes = [number_yes, number_no]
labels = ["Delivery Requests","No Delivery Requests"]
plt.pie(sizes,labels=labels,autopct="%1.1f%%")
plt.legend()
plt.show
```

# Business with high demand for delivery

```python
high_request = db[db["frequency"] == "high"]
high_request
```

#  Business with delivery request and also open for advertisement on website

```python
paying_customers = customers.fillna(0)
paying_customers = paying_customers[paying_customers["email"] !=  0]
paying_customers
```

# Business with no request yet but open for advertisement 

```python
no_paying_customers = no_customers.fillna(0)
no_paying_customers = no_paying_customers[no_paying_customers["email"] !=  0]
no_paying_customers
```

# Plot of delivery  request frequency

```python
# Number of low frequency
low = db[db['frequency'] == 'low'].count()
number_low = low['frequency']
number_low
```

```python
# Number of medium frequency
medium = db[db['frequency'] == 'medium '].count()
number_medium = medium['frequency']
number_medium
```

```python
# Number of high frequency
high = db[db['frequency'] == 'high'].count()
number_high = high['frequency']
number_high
```

```python
sizes = [number_low, number_medium, number_high ]
labels = ["Low frequency","Medium frequency", "High frequency"]
plt.pie(sizes,labels=labels,autopct="%1.1f%%")
plt.legend()
plt.show
```
